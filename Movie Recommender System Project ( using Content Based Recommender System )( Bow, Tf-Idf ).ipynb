```python
import ast                   ### ast will extract dic from string
```

```python
def sep(y):
    z=[]
    for i in range(len(ast.literal_eval(y))):
        z.append(ast.literal_eval(y)[i]['name'])
    
    return z
```

```python
def actor(y):
    v=[]
    c=0
    for i in range(len(ast.literal_eval(y))):     ## range(3)
        if c!=3:
            v.append(ast.literal_eval(y)[i]['name'])
            c+=1
        else:
            break
    return v
```

```python
new_df['cast'] = new_df['cast'].apply(actor)
```

```python
def dir(c):
    d=[]
    for i in ast.literal_eval(c):
        if i['job']=='Director':
            d.append(i['name'])
            break
    return d
```

### <font color='Red'> Remove Similar words from tag </font>

```python
### But there is problem like of same word repeat like action and actions, actor and actors

from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
```

```python
ps.stem('actors')
```

```python
def stem(v):
    y=[]
    for i in v.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new['tag'] = new['tag'].apply(stem)
```

```python
new['tag'][0]
```

# <font color='Orange'> Movies Recommendations </font>

```python
def recom(movie):
    movie_index = new[new['title']==movie].index[0]
    distance = sim_dis[movie_index]
    movies_list = sorted(list(enumerate(distance)),reverse=True,key = lambda x : x[1])[1:10]
    
    for i in movies_list:
        print(new.iloc[i[0]].title)            ## give movie title
       
```
